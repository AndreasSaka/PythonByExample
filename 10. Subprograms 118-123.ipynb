```python
#Challenge 119


import random

min_num = int(input("Enter a low number: "))
max_num = int(input("Enter a high number: "))
def guessing():
    print('I am thinking of a number: ')
    guessed = int(input('Guess the number: '))
    return guessed


def check(comp_num,guessed):
    while True:
        if guessed == comp_num:
            print('Correct you win')
            break
        elif guessed > comp_num:
            print('Oh no that is too high')
            guessed = int(input('Guess the number: '))
        elif guessed < comp_num:
            print('Oh no that is too low')
            guessed = int(input('Guess the number: '))
        
        
def main(min_num,max_num):
    comp_num = random.randint(min_num,max_num)
    guessed = guessing()
    check(comp_num,guessed)
    
if __name__ == '__main__':
    main(min_num,max_num)
    
    
```

```python
#Challenge 120


def summ():
    num1 = random.randint(5,20)
    num2 = random.randint(5,20)
    print(num1)
    print(num2)
    sumings = num1 + num2
    print(num1 + num2)
    sumoftwo = int(input("Enter your answer num1 + num2: "))
    print(f'Your answer is {sumoftwo}')
    return sumings, sumoftwo

def sub():
    num1 = random.randint(25,20)
    num2 = random.randint(1,25)
    print(num1)
    print(num2)
    subtractings = num1 - num2
    print(num1 - num2)
    suboftwo = int(input("Enter your answer num1 - num2: "))
    print(f'Your answer is {suboftwo}')
    return subtractings, suboftwo
    
def checking(x,y):
    if x == y:
        print('Correct')
    else:
        print('Incorrect')

        
def main():
    print('1) Addition')
    print('2) Subtraction')
    print('Enter 1 or 2')
    answer = input('Enter your choice: ')
    if answer == '1':
        x,y = summ()
        checking(x,y)
    elif answer == '2':
        x,y = sub()
        checking(x,y)
        
main()
        
```

```python
#Challenge 123
def create_file():
    file = open('Salaries.csv','w')
    name = input('Enter your name: ')
    salary = input('Enter your salary: ')
    entry = name + ',' + str(salary) + '\n'
    file.write(str(entry))
    file.close()

def read_file():
    file = open('Salaries.csv','r')
    print(file.read())
    file.close()
    
def delete_entry():
    list_temp = []
    file = open('Salaries.csv','r')
    for row in file:
        list_temp.append(row)
    file.close()
    print(list_temp)
    #num = int(input('Enter the number of the element you wish to change: '))
    addname = int(input("Enter the number of the element you wish to delete: "))
    del list_temp[addname]
    file = open('Salaries.csv','w')
    for row in list_temp:
        file.write(row)
    file.close()
    
    
    
def main():
    print('1) Add to file')
    print('2) View all records')
    print('3) Delete element from row')
    print('4) Quit program')
    answer = input('Enter the number of your selection: ')
    while answer != '4':
        if answer == '1':
            create_file()
            answer = input('Enter the number of your selection: ')
        elif answer == '2':
            read_file()
            answer = input('Enter the number of your selection: ')
        elif answer == '3':
            delete_entry()
        elif answer == '4':
            sys.exit()
        else:
            print('Incorrect option')
            answer = input('Enter the number of your selection: ')
            
main()
```
